````python
import json
import re

tmp = "```json\n[\n  [1, 4, 2, 2, 3, 3],\n  [2, 5, 1, 1, 6, 6, 7, 7],\n  [3, 3, 0, 0, 2, 2, 4, 4, 5, 5],\n  [4, 8, 4, 4, 1, 1, 3, 3, 5, 5, 7, 7, 9, 9],\n  [5, 10, 5, 5, 2, 2, 4, 4, 6, 6, 8, 8, 10, 10, 12, 12, 14, 14, 16, 16]\n]\n```"
print(tmp)
def extract_testcases(raw_text):
    # 去掉 ```json 和 ```
    json_text = re.sub(r"```json|```", "", raw_text).strip()
    # 解析为 Python 对象
    json_text = json_text.replace("'", '"')
    return json.loads(json_text)
print(extract_testcases(tmp))
````
